```python
%pylab inline  
import GPy
import GPyOpt
import numpy as np
import warnings
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from numpy.random import seed
from scipy.stats import multivariate_normal as mvn

warnings.filterwarnings('ignore')
seed(1996)
```

# Hyperparameter optimisation of a Gaussian process

This notebook fits a Gaussian process to precipitation data from the UKESM. It then runs hyperparameter optimisation using `GPyOpt`, an open-source global optimisation library.

The optimisation follows the structure from this [GPyOpt tutorial](https://nbviewer.org/github/SheffieldML/GPyOpt/blob/master/manual/GPyOpt_scikitlearn.ipynb).

Table of contents:

> [1. Load the data](#1.-Load-the-data)

> [2. Fit a GP to the data](#2.-Fit-a-GP-to-the-data)

> [3. Optimise the hyperparameters](#3.-Optimise-the-hyperparameters)

## 1. Load the data

We've previously exported data from a historical experiment run on the UKESM to avoid having to fetch it from the CMIP6 catalogue. Our dataset contains 1980 monthly precipitation measurements at one geographical location, latitude 51.875 and longitude 0.9375. 

```python
pr_df = pd.read_csv("../data/51_875x0_9375.csv")
```

For the experiment below, we use the first 80 months of data (starting January 1850) to fit the model, and test on the following 20 months.

```python
training_set = pr_df.iloc[0:80]
test_set = pr_df.iloc[80:100]

x_train = np.linspace(1, 80, 80)
x_test = np.linspace(80, 100, 20)

# Renormalise precipitation measurements since they're all approximately *10^-5
y_train = (training_set["pr"].to_numpy())*10**5
y_test = test_set["pr"].to_numpy()*10**5
```

## 2. Fit a GP to the data

Choosing a good kernel has an outsized influence on the quality of the GP's predictions. Let's use a covariance function that combines three kernels: a radial basis function (RBF), a standard periodic, and a periodic Matérn.

Each of these has some hyperparameters:
- RBF: length scale and variance;
- standard periodic: length scale;
- periodic Matérn: length scale and variance;

For the initial run, these are hand-picked to give reasonable performance based on previous experiments. 

```python
kernel = GPy.kern.RBF(1, lengthscale=2.6, variance=3.5) + GPy.kern.StdPeriodic(1, lengthscale=1.0) * GPy.kern.PeriodicMatern32(1, lengthscale=10, variance=3.5)
model = GPy.models.GPRegression(x_train.reshape(-1, 1), y_train.reshape(-1, 1), kernel=kernel, normalizer= True, noise_var=0.05)
```

```python
display(model)
```

```python
x_all = np.linspace(1, 100, 100)
y_mean, y_std = model.predict(x_all.reshape(-1, 1))
```

```python
sns.set()
plt.figure(figsize=(10, 5), dpi=100)
plt.xlabel("time")
plt.ylabel("precipitation")
plt.scatter(x_train, y_train, lw=1, color="b", label="training dataset")
plt.scatter(x_test, y_test, lw=1, color="r", label="testing dataset")
plt.plot(x_all, y_mean, lw=3, color="g", label="GP mean")
plt.fill_between(x_all, (y_mean + y_std).reshape(y_mean.shape[0]), (y_mean - y_std).reshape(y_mean.shape[0]), facecolor="b", alpha=0.3, label="confidence")
plt.legend(loc="upper left")
plt.show()
```

## 3. Optimise the hyperparameters

The idea behind hyperparameter optimisation is that we would like a principled way to search for values of hyperparameters that give us good results. 

One relatively simple way to do this is to do a grid search: just specify some values that each hyperparameter can take on, and then try all the combinations of those values to see which works best.

Another way is Bayesian optimisation. 

Let's consider our model — the GP we've just fit to the precipitation data — to be a black-box function  $f(\pmb \theta) = \ell$, where $\pmb \theta$ is a vector of our hyperparameters, and $\ell$ is a measure of the goodness-of-fit of the model to the precipitation data. 
We can use another Gaussian process to fit this function so that we determine the best $\pmb \theta$.

```python
# Make some predictions by drawing from the distributions 
# the GP has returned for all values (the first 80 were
# actually in the training set, the rest only test set.)
def draw_samples(y_mean, y_std):
    mean = y_mean.reshape(-1)
    cov = np.diag(y_std.reshape(-1))
    y_pred = mvn.rvs(mean, cov, 1)

    return y_pred
```

```python
def fit_gp(x):
    # x are the hyperparameters: ls1, ls2, v1, v2.
    
    rmse = 0
    for i in range(x.shape[0]):
        kernel = GPy.kern.RBF(1, lengthscale=x[i, 0], variance=x[i, 3]) + GPy.kern.StdPeriodic(1, lengthscale=x[i, 1]) * GPy.kern.PeriodicMatern32(1, lengthscale=x[i, 2], variance=x[i, 4])
        model = GPy.models.GPRegression(x_train.reshape(-1, 1), y_train.reshape(-1, 1), kernel=kernel, normalizer= True, noise_var=0.05)
        y_mean, y_std = model.predict(x_all.reshape(-1, 1))
        y_pred = draw_samples(y_mean, y_std)
        
        # Calculate RMSE — does it even make sense to use RMSE since
        # a GP doesn't make exact predictions, but specifies distributions
        # over functions?
        all_y = np.concatenate([y_train, y_test])
        rmse += np.sqrt(np.square(all_y - y_pred).mean())

    return rmse
```

Before we can run optimisation, we need to specify over which domain the optimisation should occur, for example we would like the first length scale to be $\in [0, 5]$.

```python
domain = [
    {'name': 'lengthscale1', 'type': 'continuous', 'domain': (0.,5.)},
    {'name': 'lengthscale2', 'type': 'continuous', 'domain': (0.,1.)},
    {'name': 'lengthscale3', 'type': 'continuous', 'domain': (0.,5.)},
    {'name': 'variance1', 'type': 'continuous', 'domain': (0.,1.)},
    {'name': 'variance2', 'type': 'continuous', 'domain': (0.,1.)}]
```

Of course, the GP we use for hyperparameter optimisation has its own hyperparameters. The `GPyOpt` library also lets us choose from different acquisition functions, model types, sampling types, and so on.

```python
opt = GPyOpt.methods.BayesianOptimization(f = fit_gp,            # function to optimize       
                                          domain = domain,         # box-constraints of the problem
                                          acquisition_type ='LCB',       # LCB acquisition
                                          acquisition_weight = 0.1)   # Exploration-exploitation
```

```python
opt.run_optimization(max_iter=20)
opt.plot_convergence()
```

After running Bayesian optimisation on our hyperparameters, let's look at what the optimal values seem to be:

```python
optimal_hparams = opt.X[np.argmin(opt.Y)]
print(f"ls1: {str(optimal_hparams[0])}, ls2: {str(optimal_hparams[1])}, ls3: {str(optimal_hparams[2])}, v1: {str(optimal_hparams[3])}, v2: {str(optimal_hparams[4])}\n")
```

Now, to compare the goodness-of-fit of the optimised model vs. the original model, let's plot the same graph as before.

```python
kernel = GPy.kern.RBF(1, lengthscale=optimal_hparams[0], variance=optimal_hparams[3]) + \
             GPy.kern.StdPeriodic(1, lengthscale=optimal_hparams[1]) * \
             GPy.kern.PeriodicMatern32(1, lengthscale=optimal_hparams[2], variance=optimal_hparams[4])

model = GPy.models.GPRegression(x_train.reshape(-1, 1), y_train.reshape(-1, 1), kernel=kernel, normalizer=True,
                                noise_var=0.05)
y_mean, y_std = model.predict(x_all.reshape(-1, 1))
```

```python
plt.figure(figsize=(10, 5), dpi=100)
plt.xlabel("time")
plt.ylabel("precipitation")
plt.scatter(x_train, y_train, lw=1, color="b", label="training dataset")
plt.scatter(x_test, y_test, lw=1, color="r", label="testing dataset")
plt.plot(x_all, y_mean, lw=3, color="g", label="GP mean")
plt.fill_between(x_all, (y_mean + y_std).reshape(y_mean.shape[0]), (y_mean - y_std).reshape(y_mean.shape[0]), facecolor="b", alpha=0.3, label="confidence")
plt.legend(loc="upper left")
plt.show()
```

Two things have changed:
- the mean function oscillates more, which should fit the data better since our data isn't on a smooth curve;
- the confidence bounds are tighter;

This means that if we sampled from the resulting distributions, we are more likely to get points that correspond to actual test set measurements.
