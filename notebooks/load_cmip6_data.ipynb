# Understanding the data

In this data, there is one entry for each month's precipitation, for each combination of latitude and longitude.

The latitude varies from -89.375 to 89.375 in 1.25 degree increments (143 values in total).

The longitude varies from 0.975 to 359.0625 in 1.875 degree increments (192 values times 2 bands, 384 in total).

```python
df.info()
```

For each degree of latitude, we have 760,320 measurements, because there are 1980 precipitation measurements for each of the 384 values of longitude.

One precipitation measurement per month, for 1980 months ranging from 01/1850 to 12/2014:

# Fit GP 

The idea here is to take the measurements at one specific point: lat: 51.875 lon: 0.9375 (somewhere in the UK)
The training data will the first 36 points and we will test on the following 14.
This is just to see what we get as a preliminary result from which we can then expand

```python
keys = training_set['pr'].keys()
tuples = []
for element in keys:
    if element[0] == 51.875 and element[2] == 0.9375 and element[1]==0:
        tuples.append(element)
```

```python
#Create Train/Test Train 1850-1854 Test 1855
import numpy as np
import GPy
Xtrain = np.arange(0, 36, 1)
Ytrain = np.array(training_set['pr'][tuples[:36]])*10**5
Xtest = np.arange(0, 50, 1)
Ytest = np.array(training_set['pr'][tuples[0:50]])*10**5
```

The Kernel chosen is RBF + PeriodicMatern32 - with some parameters that were tested out and showed promising results but are by no means optimized.
Commented out are model constraints and optimization function that we can use if we fiigure out how to get them to work properly

```python
kernel = GPy.kern.RBF(1, lengthscale=2.6, variance=3.5) + GPy.kern.PeriodicMatern32(1, lengthscale=10, variance=3.5)
model = GPy.models.GPRegression(Xtrain.reshape(-1, 1), Ytrain.reshape(-1, 1), kernel=kernel, normalizer= True, noise_var=0.05)
#model['rbf.variance'].constrain_bounded(3,5)
#model['rbf.lengthscale'].constrain_bounded(2,3)
#model['Gaussian_noise.variance'].constrain_bounded(0,0.5)
#model.optimize()
```

```python
display(model)
model.plot(figsize=(5, 5), dpi=100, xlabel="time", ylabel="precipitation")
```
